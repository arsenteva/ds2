**Цель проекта** — разобраться в поведении пользователей Яндекс Афиши и понять, какие из них с большей вероятностью возвращаются на платформу и покупают билеты повторно. Это нужно, чтобы быстрее находить перспективных клиентов, точнее сегментировать аудиторию, персонализировать предложения, лучше настраивать рекламу и в итоге эффективнее тратить маркетинговый бюджет и повышать удержание.

В работе используются данные о заказах и мероприятиях (выгрузка из PostgreSQL через SQL). Дальше данные приводятся в аккуратный вид: чистятся даты и пропуски, проверяются аномалии, выручка конвертируется в единую валюту (рубли), собираются нужные признаки. Затем формируется профиль пользователя — агрегированный набор характеристик (что купил в первый раз, с какого устройства, в каком регионе, через какого билетного оператора, средний чек/выручка, число билетов, временные параметры и т. п.)

После этого сравниваются пользователи с одной покупкой и с повторными, оценивается, как на вероятность возврата влияют характеристики первого заказа, состав заказа, средняя выручка, временные факторы, а также делается корреляционный анализ между признаками профиля и количеством покупок. По итогам формулируются выводы и практические рекомендации для маркетинга. Финальная часть — оформление результата и публикация проекта в Git.

````python
load_dotenv()

db_config = {
    'user': os.getenv('DB_USER'),      # имя пользователя
    'pwd': os.getenv('DB_PASSWORD'),   # пароль
    'host': os.getenv('DB_HOST'),      # хост
    'port': os.getenv('DB_PORT'),      # порт
    'db': os.getenv('DB_NAME')         # название базы данных
}

connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
) 

engine = create_engine(connection_string) 

query = '''
SELECT p.user_id,
        p.device_type_canonical,
        p.order_id,
        p.created_dt_msk AS order_dt,
        p.created_ts_msk AS order_ts,
        p.currency_code,
        p.revenue,
        p.tickets_count,
        p.created_dt_msk::date - LAG(created_dt_msk::date) OVER (
          PARTITION BY user_id
          ORDER BY created_dt_msk::date
        ) AS days_since_prev,
        e.event_id,
        e.event_name_code AS event_name,
        e.event_type_main,
        p.service_name,
        r.region_name,
        c.city_name
FROM afisha.purchases AS p 
INNER JOIN afisha.events AS e ON p.event_id = e.event_id
INNER JOIN afisha.city as c ON e.city_id = c.city_id
INNER JOIN afisha.regions as r ON c.region_id = r.region_id
WHERE device_type_canonical IN ('mobile', 'desktop') 
AND e.event_type_main != 'фильм'
ORDER BY user_id
''' 
````

````python
# Объединим датасеты по полю даты

df = df.merge(df_tenge[['one_tenge_curs', 'data']], left_on='order_dt', right_on='data')
df.info()
````

````python
# Высчитаем данные для нового столбца revenue_rub

def add_revenue_rub(row):
    if row['currency_code'] == 'kzt':
        return row['revenue'] * row['one_tenge_curs']
    else:
        return row['revenue']
    
df['revenue_rub'] = df.apply(add_revenue_rub, axis=1)
````
